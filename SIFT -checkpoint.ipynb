```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
read=cv2.imread("pigeon1.jpg")
read=cv2.resize(read,(250,250))
```

```python
read1=cv2.imread("pigeon2.jpg")
read1=cv2.resize(read1,(250,250))
```

```python
read=cv2.cvtColor(read,cv2.COLOR_BGR2GRAY)
```

```python
read1=cv2.cvtColor(read1,cv2.COLOR_BGR2GRAY)
```

```python
sift=cv2.SIFT_create()
```

```python
keypoints_1,descriptors_1=sift.detectAndCompute(read,None)
keypoints_2,descriptors_2=sift.detectAndCompute(read1,None)
bf=cv2.BFMatcher(cv2.NORM_L1,crossCheck=True)
matches=bf.match(descriptors_1,descriptors_2)
matches=sorted(matches,key=lambda x:x.distance)
read3=cv2.drawMatches(read,keypoints_1,read1,keypoints_2,matches[:100],read,flags=2)
plt.imshow(read3),plt.show()
```
