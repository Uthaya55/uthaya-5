```python
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
import matplotlib.pyplot as plt
```

# READING THE IMAGE

```python
read=imread("jack.jpg")
```

```python
plt.axis("off")
plt.imshow(read)
read.shape
```

```python
read=resize(read,(128*10,64*10))
plt.axis("off")
plt.imshow(read)
read.shape
```

```python
fd,hog_image=hog(read,orientations=10,pixels_per_cell=(2,2),cells_per_block=(4,4),visualize=True,channel_axis=2)
fd.shape
hog_image.shape
plt.axis("off")
plt.imshow(hog_image,cmap='Blues_r')
plt.show()
```
